# Depth-First Search (DFS)
* Uses a **stack** data structure (LIFO - Last In First Out)
* Explores as far as possible along each branch before backtracking
* Goes deep into the graph/tree before exploring siblings
* May not find the optimal/shortest path
* Uses less memory than BFS
* Good for: topological sorting, detecting cycles, finding connected components
* Time Complexity: O(V + E) where V = vertices, E = edges
* Space Complexity: O(V) - for the stack

# Breadth-First Search (BFS)
* Uses a **queue** data structure (FIFO - First In First Out)
* Will always find the optimal path but requires a lot of memory
* Explores all neighbors at the current depth level before moving deeper
* Explores level-by-level or layer-by-layer
* Guarantees shortest path in unweighted graphs
* Uses more memory than DFS
* Good for: finding shortest path, level-order traversal, social networks
* Time Complexity: O(V + E) where V = vertices, E = edges
* Space Complexity: O(V) - for the queue (can be larger than DFS in practice)
