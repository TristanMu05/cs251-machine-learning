Lists in python (mutability)
 * list data structure can be changed after being created
 * multiple variable names can point to the same object in memory

```python
mya = [4,5,6]
myb = mya
myb.append(7)
print(mya)
print(myb)
```

How to make a class in python:

1. Define the class
2. Initialize the class
3. Create an instance of the class
4. Use the instance


```python
class dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print("Woof!")

def main():
    my_dog = dog("Buddy")
    my_dog.bark()

if __name__ == "__main__":
    main()
```
